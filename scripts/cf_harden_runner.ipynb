```python
# Step 1: Set security_level to 'low'
import subprocess, os
os.chdir(r'c:\Users\sossi\Desktop\Orchestrator Hedge Edge')
env = os.environ.copy()
env['PYTHONIOENCODING'] = 'utf-8'

print("=" * 60)
print("STEP 1 â€” Set security_level to 'low'")
print("=" * 60)
result1 = subprocess.run(
    [r'.venv\Scripts\python.exe', '-c',
     "from shared.cloudflare_client import set_security_level, ZONE_ID; "
     "result = set_security_level(ZONE_ID, 'low'); "
     "print('Security level set to:', result.get('value', result))"],
    capture_output=True, timeout=60, encoding='utf-8', errors='replace', env=env
)
print("STDOUT:", result1.stdout.strip())
if result1.stderr.strip():
    print("STDERR:", result1.stderr.strip())
print(f"Exit code: {result1.returncode}")

print()
print("=" * 60)
print("STEP 2 â€” Hardening dry-run verification")
print("=" * 60)
result2 = subprocess.run(
    [r'.venv\Scripts\python.exe', '-m', 'shared.cloudflare_harden'],
    capture_output=True, timeout=60, encoding='utf-8', errors='replace', env=env
)
print("STDOUT:", result2.stdout.strip())
if result2.stderr.strip():
    print("STDERR:", result2.stderr.strip())
print(f"Exit code: {result2.returncode}")

# Also save both outputs to a file for reference
with open('_cf_security_low_output.txt', 'w', encoding='utf-8') as f:
    f.write("STEP 1 â€” set_security_level('low')\n")
    f.write(f"STDOUT: {result1.stdout}\n")
    f.write(f"STDERR: {result1.stderr}\n")
    f.write(f"Exit: {result1.returncode}\n\n")
    f.write("STEP 2 â€” cloudflare_harden dry-run\n")
    f.write(f"STDOUT: {result2.stdout}\n")
    f.write(f"STDERR: {result2.stderr}\n")
    f.write(f"Exit: {result2.returncode}\n")
print("\nBoth outputs saved to _cf_security_low_output.txt")
```
